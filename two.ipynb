```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df_sumup = pd.read_excel("附件1.xlsx")
df_buy = pd.read_excel("buy.xlsx")
```

```python
year_result=[0,0,0]
for index, row in df_buy.iterrows():
    # 获取当前行的信息
    time = row["销售日期"]
    name = row["分类名称"]
    w = row["销量(千克)"]
    type=row["销售类型"]
    year, month_, _ = (str)(time).split("-")
    year = (int)(year)-2020
    month=(int)(month_)
    if month<7:
        year_result[year-1]+=1
    else:
        year_result[year]+=1
categories = ['20-7~21-6', '21-7~22-6', '22-7~23-6']
# 绘制柱形图
plt.rcParams['font.sans-serif'] = ['SimSun']  # 设置中文字体为宋体（根据实际字体文件名填写）
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
plt.bar(categories, year_result)

# 设置图表标题和轴标签
plt.title('三年总销量变化')
plt.xlabel('年份')
plt.ylabel('总销量')
plt.savefig('three_year_sum.png', format='png')
# 显示图表
plt.show()
```
